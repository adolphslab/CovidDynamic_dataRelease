```python
tasks = ['biat', 'iat', 'amp', 'tr_1s', 'altt', 'pgg','cvd_consp']

for wave in task_quality.wave.unique():
    for task in tasks:
        if sum(task_quality.loc[task_quality.wave == wave, task+'_administered'] == 'True')>0:
            task_quality.loc[task_quality.wave == wave, task+'_administered'] = 'True'
        else:
             task_quality.loc[task_quality.wave == wave, task+'_administered'] = 'False'

# w4 50% git biat, 50% git iat and amp. FInd out who got which
w4_biat_idx = (task_quality.wave == '4') &  (task_quality.biat_totalTime != 'nan')
task_quality.loc[w4_biat_idx, ['iat_administered','amp_administered']] = 'False'
w4_iat_amp_idx = (task_quality.wave == '4') &  ((task_quality.iat_totalTime != 'nan') | (task_quality.amp_totalTime != 'nan'))
task_quality.loc[w4_iat_amp_idx, ['biat_administered']] = 'False'

# mark missing task data
for task in tasks:
    cols = task_quality.columns[task_quality.columns.str.startswith(task)]
    cols = list(set(cols) - set([task+'_administered']))
    if task == 'altt':
        cols = list(set(cols)-set(['altt_meanLongString']))
    df_tmp = pd.DataFrame()
    df_tmp = task_quality[cols] == 'nan'
    task_quality.loc[(df_tmp.sum(axis=1)>0) & (task_quality[task+'_administered']=='True'), task+'_missing'] = 'True'
    task_quality.loc[(task_quality[task+'_missing'].isnull())& (task_quality[task+'_administered']=='True'), task+'_missing'] = 'False'

    task_quality.loc[task_quality[task+'_administered'] == 'False', [task+'_missing']] = 'False'

task_quality.to_csv('/Users/trusch/Box/COVID-19 Adolphs Lab/PreProcessed_Data/validation_files/all_tasks_w1-17_A-N.csv',index=False)

```

```python
def plotTaskHist_Conte(task,outlier_perWave, validVar):
    """
    histograms of total validation meassure in each wave
    outlier subejcts (identified beforehand) marked with red dots
    """
    fig, axs = plt.subplots(2, 3,figsize=(15, 10))
    waves = ['A','B','C','D','E']
    subplot_row = {'A': 0, 'B': 0, 'C': 0,'D': 1,'E': 1}
    subplot_col = {'A': 0, 'B': 1, 'C': 2,'D': 0,'E': 1}


    for i_wave in waves: 
        data = task_validation_conte.loc[task_validation_conte['wave']==i_wave,task+'_' + validVar]
        if not data.isnull().all():
            axs[subplot_row[i_wave],subplot_col[i_wave]].hist(list(data),bins=20)

            outlier = outlier_perWave[i_wave]
            for i_out in outlier:
                try:
                    outlier_data = data[i_out]
                    axs[subplot_row[i_wave],subplot_col[i_wave]].plot(outlier_data,1,'ro') 
                except:
                    print('wave ' + i_wave + ' ' +i_out + ' ' + task + ' does not exist')

            axs[subplot_row[i_wave],subplot_col[i_wave]].set_title('wave ' + i_wave + ' ' + task)
            axs[subplot_row[i_wave],subplot_col[i_wave]].set_xlabel(validVar, fontsize=16)  
            axs[subplot_row[i_wave],subplot_col[i_wave]].set_ylabel('subject frequency', fontsize=16)  
            fig.savefig(fig_dir + '/' + task + '_Conte_' + validVar +'.png', dpi = 300)
```
