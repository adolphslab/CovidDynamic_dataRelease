```python
def state_tscore_on_table(r):
    raw = float(r["STAI_State_raw"])

    if np.isnan(raw):
        return np.nan
    else:
        raw = int(raw)
    #print(raw)
    gender_val = float(r['DemC5'])
    #print(gender_val)
    age_val = float(r['age'])
    #print(age_val, gender_val, raw)
    #print(r['PROLIFIC_PID'])
    
    out_val = np.nan
    
    if (raw > 80) | (raw <20):
        out_val = np.nan
    else:
        if np.isnan(age_val) | np.isnan(gender_val) | np.isnan(raw) :
            out_val = np.nan
        elif ((age_val <= 39) & (age_val >=17)) & (gender_val==1):
            out_val = lookup.loc[raw][0]
        elif ((age_val <= 39) & (age_val >=17)) & (gender_val==2):
            out_val = lookup.loc[raw][2]
        elif ((age_val <= 49) & (age_val >=40)) & (gender_val==1):
            out_val = lookup.loc[raw][4]
        elif ((age_val <= 49) & (age_val >=40)) & (gender_val==2):
            out_val = lookup.loc[raw][6]
        elif ((age_val <= 71) & (age_val >=50)) & (gender_val==1):
            out_val = lookup.loc[raw][8]
        elif ((age_val <= 71) & (age_val >=50)) & (gender_val==2):
            out_val = lookup.loc[raw][10]
    return out_val

def trait_tscore_on_table(r):
    raw = float(r["STAI_Trait_raw"])

    if np.isnan(raw):
        return np.nan
    else:
        raw = int(raw)
    #print(raw)
    gender_val = float(r['DemC5'])
    #print(gender_val)
    age_val = float(r['age'])
    #print(age_val, gender_val, raw)
    
    out_val = np.nan
    
    if (raw > 80) | (raw <20):
        out_val = np.nan
    else:
        if np.isnan(age_val) | np.isnan(gender_val) | np.isnan(raw) :
            out_val = np.nan
        elif ((age_val <= 39) & (age_val >=17)) & (gender_val==1):
            out_val = lookup.loc[raw][1]
        elif ((age_val <= 39) & (age_val >=17)) & (gender_val==2):
            out_val = lookup.loc[raw][3]
        elif ((age_val <= 49) & (age_val >=40)) & (gender_val==1):
            out_val = lookup.loc[raw][5]
        elif ((age_val <= 49) & (age_val >=40)) & (gender_val==2):
            out_val = lookup.loc[raw][7]
        elif ((age_val <= 71) & (age_val >=50)) & (gender_val==1):
            out_val = lookup.loc[raw][9]
        elif ((age_val <= 71) & (age_val >=50)) & (gender_val==2):
            out_val = lookup.loc[raw][11]
    return out_val

def bdi_range(r):
    bdi = float(r['BDI_total_raw'])
    out = np.nan
    if np.isnan(bdi):
        out = np.nan
    elif bdi <= 13:
        out = 1
    elif (bdi <= 19) & (bdi >=14):
        out = 2
    elif (bdi <= 28) & (bdi >=20):
        out = 3
    elif (bdi <= 63) & (bdi >=29):
        out = 4
    
    return out

def state_trait_range(r, state_trait):
    st = float(r[state_trait])
    out = np.nan
    if np.isnan(st):
        out = np.nan
    elif (st <= 37) & (st >=20):
        out = 1
    elif (st <= 44) & (st >=38):
        out = 2
    elif (st <= 63) & (st >=45):
        out = 3
    return out
```
